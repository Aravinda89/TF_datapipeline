```python
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

value = randint(0, 45000, 1)[0]
random_index = value[0]

plt.figure(figsize=(20,20))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_imgs[random_index+i], cmap=plt.cm.binary)
    plt.xlabel(class_names[train_labels[random_index+i][0]])
plt.show()
```
